## **Behavior of Numeric Variables**

To analyze the numeric variables, the first step will be to look at their distribution and then their importance in relation to whether the transaction ends up being fraudulent or not.

First, we will observe this without graphical behavior, only by examining what is and what is not fraud with respect to each column. This is:


```python
columns_ = data.select_dtypes(include=['float64', 'int64']).columns

for i in columns_:
    fraude_no_fraude_counts = data.groupby('fraude')[i].count()
    print(f"Counts of Fraud and Not Fraud for variable '{i}':")
    print(fraude_no_fraude_counts)
    print("=====================================================")
```

If we carefully observe the output of the previous code, we see that for almost all the columns, the amount of fraud and non-fraud is exactly the same. This could indicate two behaviors: either the variables in their current state have no relationship with the target, or they may be presenting high multicollinearity among them.

This leads us to think that we need to perform thorough testing when selecting the features we want to use for the model. Relying on decision trees, RFE, mutual_information_score, and other techniques will be necessary to ensure the model remains coherent.


```python
import warnings
warnings.filterwarnings("ignore")

numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
excluded_columns = ['c', 'd', 'f', 'g', 'k', 'q', 'r', 's','monto']
filtered_columns = [col for col in numerical_columns if col not in excluded_columns]

for col in filtered_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x=col, hue='fraude', palette="Set2", edgecolor='black')
    plt.title(f'Distribución de la Variable: {col} por Tipo de Fraude')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title='Fraude', labels=['No fraude (0)', 'Sí fraude (1)'])
    plt.tight_layout()
    plt.show()
```

To continue with the analysis of numerical variables, we can use PCA. This is because PCA transforms the original variables into principal components that capture most of the variance in the dataset, allowing us to visualize the data in the first two or three principal components to observe patterns, trends, and natural groupings.

This is especially useful for understanding how the data is distributed and if there are possible groups or clusters, as is the case in our classification context.

The results of this PCA can provide baseline clues about which features to use for our model.


```python
data_copy = data.copy()
data_copy.drop(columns=['k'], inplace=True)
data_copy['c'].fillna(data_copy['c'].mean(), inplace=True)

numeric_data = data_copy.select_dtypes(include=[np.number]).drop(columns=['fraude'])
#scaler = StandardScaler()
scaler = MinMaxScaler()
numeric_data_scaled = scaler.fit_transform(numeric_data)

# PCA con 10 componentes
pca = PCA(n_components=10)
pca_components = pca.fit_transform(numeric_data_scaled)

# df con los componentes principales
pca_df = pd.DataFrame(data=pca_components, columns=[f'PC{i+1}' for i in range(10)])
pca_df['fraude'] = data_copy['fraude'].values[:len(pca_df)]

# Clustering Jerárquico para formar grupos
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
pca_df['Cluster'] = hc.fit_predict(pca_df.drop(columns=['fraude']))

# clústeres en los primeros dos componentes principales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', style='fraude', palette='viridis', markers=["o", "s"])
plt.title("Agrupación de los Componentes Principales con relación al Fraude")
plt.show()
```

This graph shows that clusters $0$, $1$, and $2$ are relatively separated in the principal component space, suggesting that PCA has allowed the identification of distinct data patterns or groups. Additionally, the graph shows that the dataset can be represented meaningfully in two principal components, though not completely, but somewhat approaching the reality of the data. However, whether the transaction is fraudulent or not is not clear for any of the clusters, indicating that further analysis and feature engineering on the variables and data are needed to obtain more accurate information.


```python
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(10)], index=numeric_data.columns)
top_10_pc1 = loadings['PC1'].abs().nlargest(10)

print("Las 10 variables que más contribuyen al primer componente principal (PC1) son:")
print(top_10_pc1)
```

This tells us that, for example, the features `s` and `b` explain a significant amount of the variance in the data, which could indicate that their influence in detecting fraud may be high.
