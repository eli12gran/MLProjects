The places on the Titanic were divided as follows:

![](https://i.imgur.com/FGUPMgl.png)

This visual corresponds to the section of the center of the ship. The ship was marked by letters, except for the upper deck ‚Äîalso known as the boat deck‚Äî and the two lower decks dedicated to machinery. The highest deck was, therefore, deck A, and the letters continued down to G. **The first class extended from the upper deck to deck E**. **The RMS Titanic struck the iceberg primarily with its bow**, the front part of the ship. The collision occurred on the starboard side of the bow, damaging several watertight compartments of the ship, which eventually caused it to sink.

### **Feature Engineering**

In this part of Feature Engineering, we will first remove the columns that do not provide us with much information, such as `passengerid, ticket, and name`. In addition, we will change the `sex` variable to be numerical so that men are represented by 0 and women by 1.


```python
df_train.drop(columns=['passengerid', 'ticket'], inplace=True)
df_full_train.drop(columns=['passengerid', 'ticket'], inplace=True)
df_test.drop(columns=['passengerid', 'ticket'], inplace=True)
df_val.drop(columns=['passengerid', 'ticket'], inplace=True)
```

Now we will take care of the null data. As we could see at the top, the variables `age`, `cabin`, and `embarked` have null data, with age and cabin having the most.

For the `embarked` feature, we will fill the two null spaces with the mode.

For the `age` feature, we must be a little more careful because as we observed in the EDA in the part where we analyzed age, we could see that, for example, children had a higher probability of being saved than adults, so if we filled this variable without being careful we could make mistakes that would cause us to have very bad results in the model evaluation.

After observing several behaviors of the `age` feature with respect to others, we came to the conclusion that based on the features `sibsp` and/or `parch`, we could fill the data better than if we used the mean age by gender (`sex`) or by `pclass`. With the latter, the box and whisker plots showed that there were many outliers, and with the gender variable, we would be biasing the information a lot.

Let's see then the information about the ages that the features `sibsp` and `parch` give us:

What we observe in these box and whisker plots about both features is that, for example, there are not many outliers, as was the case when testing with other features. **To fill the empty spaces of the `age` feature, we will finally choose the mean of the ages by the number of parents/children on board that a passenger had, that is, using the `parch` feature.**

I chose this feature and not the other one more for a matter of intuition: it seems to me that it makes more sense that, for example, the average age of a person who carried 4 companions on board was approximately 44 years rather than approximately 8 years. In addition to this, if I filled the age with these very low values, I would be assuming that most of the 177 NaN data that the age contained were children, which would imply that they would have a fairly high probability/chance of being saved and if we observe the code cell that is just above this text, we can see that children were almost five times less than adults, so that is why I considered it better to use the average ages of the `parch` feature than those of the `sibsp` feature.

Observing the dataset a bit, another idea occurred to me to fill the null values of the variable age. I don't know if it's correct like the previous one, but my intuition tells me that both could work well.


The second idea I had to fill the null values of the feature `age` was using the information that the names brought. When we look at the data we see that there are some behaviors with respect to the ages and names of the people. In the English language, married people are written as ***Mr*** if they are a married man and ***Mrs*** if they are a married woman. In the dataset, we have that, for example, married women have, for example: ***Mrs ..... (real name of the woman)***. This means that we are being given information about whether the passenger is married or not and therefore **the age of the person will be according to their marital status!**.

If we remember that quite wealthy people traveled on the Titanic and that it was also the year 1912 and there were still "titles" or prefixes for people who had money, this can then give us information about the ages of the other people. So if we extract the average age for each of the titles that people have, we would have a more appropriate way to fill in the data. I know this explanation is not very clear, so let's see what I mean here:

```python
df_full_train_copy = df_full_train.copy()
df_full_train_copy["person_title"] = df_full_train_copy["name"].str.extract(r',\s*([^\.]+)\.')
average_age_by_title = df_full_train_copy.groupby("person_title")["age"].mean()
average_age_by_title
```

Let's observe that the age of women who have ***Mrs*** in their name, have on average approximately 37 years, which makes a lot of sense: a woman so young is not very likely to be married. If we look at the title ***master*** we can see that the average age is approximately 4 years. This makes a lot of sense as well since in the 1900s, the term "Master" was used as a title for boys, especially for the sons of families of a certain social position. It was a title of respect, similar to "Mr." for adult men. Generally, it was used for boys until they reached adolescence.

```python
# rellenamos los datos emabarked con la moda
df_train.embarked = df_train.embarked.fillna(df_train.embarked.mode()[0])
df_full_train.embarked = df_full_train.embarked.fillna(df_full_train.embarked.mode()[0])
df_test.embarked = df_test.embarked.fillna(df_test.embarked.mode()[0])
df_val.embarked = df_val.embarked.fillna(df_val.embarked.mode()[0])
```

```python
# rellenamos age usando parch
def fill_age(row):
    if np.isnan(row['age']):
        return average_age_by_parch.get(row['parch'], np.nan)
    return row['age']

df_train['age'] = df_train.apply(fill_age, axis=1).round(0)
df_full_train['age'] = df_full_train.apply(fill_age, axis=1).round(0)
df_test['age'] = df_test.apply(fill_age, axis=1).round(0)
df_val['age'] = df_val.apply(fill_age, axis=1).round(0)
```

**Decision Tree**

In order to have everything in numerical terms, we will use Label Encoder for the `embarked` feature in order to assign a number to each port of embarkation. And we will use One Hot Encoding for the `sex` variable, so that we have `sex_male` and it is 1 if it is a man and 0 if it is a woman. This leaves us with:

```python
def One_Hot_Encoding(df, column, encoder=None):
    if encoder is None:
        encoder = OneHotEncoder(drop='first', sparse_output=False)
        encoded_data = encoder.fit_transform(df[[column]])
    else:
        encoded_data = encoder.transform(df[[column]])

    df_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out([column]))
    df_final = pd.concat([df, df_encoded], axis=1)
    df_final = df_final.drop(columns=[column], axis=1)

    return df_final, encoder
```

```python
%%capture
df_full_train, encoder = One_Hot_Encoding(df_full_train, "sex")
df_train, _ = One_Hot_Encoding(df_train, "sex", encoder)
df_test, _ = One_Hot_Encoding(df_test, "sex", encoder)
df_val, _ = One_Hot_Encoding(df_val, "sex", encoder)
```

## **Conclusions**

Before starting with the conclusions about the performance of the models, it is important to know that in this Titanic dataset the data was unbalanced but not as much as the one we have studied on previous occasions, so I considered accuracy to be an important metric to include in the results. However, the most important metrics in this case are **recall and F1-score**. In this context, recall gives us information about the people who survived the Titanic and the model is effectively predicting them as survivors. And a high F1-Score indicates that the model is performing well both in identifying survivors and in not over-predicting survival in those who did not.

* Of the six models implemented, I consider that **the best results in terms of generalization and adequacy come from the logistic regression and decision tree models**. When comparing the prediction results with the test data, these models have the highest scores in recall and F1-score. Although they are not the highest in the validation data, the F1-score, for example, is not significantly below the results obtained with the more complex models. This suggests that they offer a competitive performance without the additional complexity.
    
    In addition to the results in these metrics, I believe that these models offer a better result because, when observing the histograms of the probability distribution (which show the scores and the survivors), it can be seen that the classification between surviving or dying is not as polarized as it occurs with models based on Boosting techniques. This behavior indicates that the models are generalizing better and may be an indication that it is not falling into overfitting. Something similar is observed in the ROC curve: a curve that is too "perfect" could be an indication of overfitting to the training data, which would affect the model's ability to predict correctly with new data. This combination of balanced metrics and more robust behavior in generalization suggests that logistic regression and decision trees are more suitable options compared to the other models.

* I consider that although it is known how good and complex the models based on Boosting techniques and Random Forest models are, these models did not provide very good results mainly due to the size of the dataset: the robustness of these models generally requires a large volume of data, so that it does not fall into overfitting errors. In addition to this, it is also possible that the new features we added (such as `has cabin`) are not providing enough information for complex models to take full advantage of their capabilities. Instead, simple models can more efficiently take advantage of the features that were selected, especially if these features are already informative on their own.

* One of the biggest advantages of the fact that logistic regression and decision tree models work very well in this dataset is that they are relatively simpler than the others we used and tend to be more interpretable and less susceptible to variations in the data. This can translate into better performance in situations where interactions between variables are not very complex or when additional features do not add much predictive value.

* Leaving aside the fact that the results of Logistic Regression and decision trees I believe give us more generality in predicting survival or not, of the Boosting-based models that were tested, I believe that although they all had very similar results in the metrics, the one that provided the best result not only based on the metric values but also on the observation of the ROC curve, was the **AdaBoost** model. This is because if we look at the tables at the top, in the comparison of the prediction with the validation data, AdaBoost obtained the best results for recall and F1-score, which are the metrics that matter most to us. Also, if we look at the ROC curve, we see that it has a behavior more similar to the ROC curve of logistic regression and decision trees, where we can conclude that it is not falling much into overfitting.
    
    However, if we observe the probability distribution graph for this model and also for GradientBoost, XGBoost and Random Forest, they are all very very polarized, that is, they show that the models are very sure about the decisions they are making, that they predict very correctly the people who die and live. This may be happening because boosting techniques minimize and penalize much more strongly classification errors. This is the opposite and/or contrary to what happens in simpler models like logistic regression: this predicts more smoothed probabilities because it is a linear model. It does not make hard cuts like a decision tree. Therefore, the predicted probabilities tend to be distributed more uniformly, and do not tend so much towards the extremes.

* In the Feature Engineering part, I gave two ideas to fill the null data of `age` since I considered (and the decision tree verified) that it was an important feature to decide on survival or not on the ship. However, after testing the six classification models with both ideas, based on the results of the metrics, I can say that filling the age with the `parch` variable gives better predictions: the metrics decrease their value a little in almost all models filling the age based on the name. I'm not sure if filling the null data of the age with `parch` was the best option, but I'm trusting my gut. I leave both ideas written in case the other was better :))).
